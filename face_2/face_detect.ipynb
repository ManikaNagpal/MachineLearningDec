```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import os
```

```python
#loading all directories and name of folder as label to every image inside
files = os.listdir('./data/')   #current directory
faces_paths = []
labels = []

for f in files:
    ims = os.listdir('./data/' + f)
    for px in ims:
        faces_paths.append('./data/' + f + '/' + px)
        labels.append(f)

print(faces_paths)
print(labels)
```

```python
imgs=[]

for fx in faces_paths:
    im=cv2.imread(fx)
    im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    imgs.append(im)
   
print(len(imgs))
print(imgs)
```

```python
all_faces=[]
plt.figure(0)
for mm in range(5):
    im=imgs[mm][100:500,550:850,:]       #crop image [height,length,all colors]
    print(im.shape)
    img=cv2.resize(im,(100,100))          #resolution changed to 100*100
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)    #convert color to gray
    print(img.shape)
    all_faces.append(img)        #addd to all_faces
    plt.imshow(img,cmap='gray')
    plt.show()
    
for mm in range(6,12):             #same for second folder images
    im=imgs[mm][100:590,500:800,:]
    print(im.shape)
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    all_faces.append(img)
    print(img.shape)
    plt.imshow(img,cmap='gray')
    plt.show()    
```

```python
# print(len(all_faces))      #convert all_faces to array

print(all_faces)
all_faces = np.asarray(all_faces)
print("--------------------------------------")
# print(all_faces.shape)
print(all_faces)
```

```python
lab = np.asarray(labels[:])          #save all labels and facedata to npy files
print(lab)
print(lab.shape)

np.save('./facelabels', lab)
np.save('./facedata', all_faces)
```
